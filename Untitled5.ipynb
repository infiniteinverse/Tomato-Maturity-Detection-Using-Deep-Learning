```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms
from PIL import Image
from timm.models.vision_transformer import VisionTransformer
import timm
from tqdm import tqdm
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Custom Attention Block
class EnhancedAttention(nn.Module):
    def __init__(self, dim, num_heads=8):
        super().__init__()
        self.num_heads = num_heads
        head_dim = dim // num_heads
        self.scale = head_dim ** -0.5

        self.qkv = nn.Linear(dim, dim * 3)
        self.proj = nn.Linear(dim, dim)

        # Novel addition: Attention gating mechanism
        self.gate = nn.Sequential(
            nn.Linear(dim, dim // 4),
            nn.GELU(),
            nn.Linear(dim // 4, dim),
            nn.Sigmoid()
        )

        # Novel addition: Spatial attention
        self.spatial_attn = nn.Sequential(
            nn.Conv2d(1, 1, kernel_size=5, padding=2),
            nn.Sigmoid()
        )

    def forward(self, x):
        B, N, C = x.shape
        qkv = self.qkv(x).reshape(B, N, 3, self.num_heads, C // self.num_heads).permute(2, 0, 3, 1, 4)
        q, k, v = qkv[0], qkv[1], qkv[2]

        attn = (q @ k.transpose(-2, -1)) * self.scale
        attn = attn.softmax(dim=-1)

        x = (attn @ v).transpose(1, 2).reshape(B, N, C)

        # Apply gating
        gate = self.gate(x)
        x = x * gate

        # Reshape for spatial attention (assuming we have a square input)
        seq_len = int(np.sqrt(N))
        if seq_len**2 == N:  # Ensure we have a perfect square for reshaping
            x_spatial = x[:, 1:, :].mean(dim=-1).reshape(B, 1, seq_len, seq_len)
            spatial_weight = self.spatial_attn(x_spatial)
            x_spatial = x[:, 1:, :].reshape(B, seq_len, seq_len, C)
            x_spatial = x_spatial * spatial_weight.unsqueeze(-1)
            x = torch.cat([x[:, 0:1, :], x_spatial.reshape(B, N-1, C)], dim=1)

        return self.proj(x)

# Custom MLP Block
class EnhancedMLP(nn.Module):
    def __init__(self, in_features, hidden_features, out_features):
        super().__init__()
        self.fc1 = nn.Linear(in_features, hidden_features)
        self.act = nn.GELU()
        self.fc2 = nn.Linear(hidden_features, out_features)

        # Novel addition: Skip connection with gating
        self.gate = nn.Sequential(
            nn.Linear(in_features, hidden_features // 4),
            nn.GELU(),
            nn.Linear(hidden_features // 4, out_features),
            nn.Sigmoid()
        )

        # Novel addition: Feature recalibration
        self.recalibration = nn.Sequential(
            nn.Linear(in_features, hidden_features // 2),
            nn.LayerNorm(hidden_features // 2),
            nn.GELU(),
            nn.Linear(hidden_features // 2, out_features)
        )

    def forward(self, x):
        # Main path
        main = self.fc1(x)
        main = self.act(main)
        main = self.fc2(main)

        # Skip connection with gate
        gate = self.gate(x)

        # Feature recalibration
        recalib = self.recalibration(x)

        # Combine
        return main * gate + recalib * (1 - gate)

# Custom Vision Transformer
class TomatoViT(nn.Module):
    def __init__(self, num_classes, pretrained_model="vit_base_patch16_224"):
        super().__init__()

        # Load pretrained ViT
        self.base_model = timm.create_model(pretrained_model, pretrained=True)

        # Get embedding dimension from base model
        embed_dim = self.base_model.embed_dim

        # Replace specific transformer blocks with our enhanced versions
        # Note: We'll replace some of the later blocks to keep early feature extraction intact
        for i in range(len(self.base_model.blocks) - 3, len(self.base_model.blocks)):
            # Replace attention with our enhanced attention
            self.base_model.blocks[i].attn = EnhancedAttention(dim=embed_dim, num_heads=self.base_model.blocks[i].attn.num_heads)

            # Replace MLP with our enhanced MLP
            mlp_dim = self.base_model.blocks[i].mlp.fc1.out_features
            self.base_model.blocks[i].mlp = EnhancedMLP(
                in_features=embed_dim,
                hidden_features=mlp_dim,
                out_features=embed_dim
            )

        # Replace the classification head
        self.head = nn.Sequential(
            nn.Linear(embed_dim, embed_dim),
            nn.LayerNorm(embed_dim),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(embed_dim, num_classes)
        )

        # Replace the original head
        self.base_model.head = self.head

    def forward(self, x):
        return self.base_model(x)

# Dataset class for tomato images
class TomatoDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
        self.class_to_idx = {cls: i for i, cls in enumerate(self.classes)}

        self.images = []
        self.labels = []

        for cls in self.classes:
            class_path = os.path.join(root_dir, cls)
            if os.path.isdir(class_path):
                for img_name in os.listdir(class_path):
                    if img_name.endswith(('.jpg', '.jpeg', '.png')):
                        self.images.append(os.path.join(class_path, img_name))
                        self.labels.append(self.class_to_idx[cls])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        label = self.labels[idx]

        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        return image, label

    def get_class_counts(self):
        counts = Counter(self.labels)
        return {self.classes[label]: count for label, count in counts.items()}

# Function to create weighted sampler for handling class imbalance
def create_weighted_sampler(dataset):
    # Count instances of each class
    class_counts = Counter(dataset.labels)

    # Calculate class weights (inverse of frequency)
    total_samples = len(dataset)
    class_weights = {cls: total_samples / count for cls, count in class_counts.items()}

    # Assign weight to each sample
    sample_weights = [class_weights[label] for label in dataset.labels]
    sample_weights = torch.DoubleTensor(sample_weights)

    # Create weighted sampler
    sampler = WeightedRandomSampler(sample_weights, len(sample_weights))
    return sampler

# Training function with focal loss option for handling class imbalance
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction
        self.ce_loss = nn.CrossEntropyLoss(reduction='none')

    def forward(self, inputs, targets):
        ce_loss = self.ce_loss(inputs, targets)
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# Training function
def train_model(model, dataloader, criterion, optimizer, device, num_epochs=10):
    model.to(device)
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        pbar = tqdm(dataloader['train'])
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

            pbar.set_description(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item():.4f}")

        train_loss = running_loss / len(dataloader['train'])
        train_acc = correct / total
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)

        # Validation
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in dataloader['val']:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                val_loss += loss.item()
                _, predicted = outputs.max(1)
                val_total += labels.size(0)
                val_correct += predicted.eq(labels).sum().item()

        val_loss = val_loss / len(dataloader['val'])
        val_acc = val_correct / val_total
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

    return history

# Function to evaluate model and create confusion matrix
def evaluate_model(model, dataloader, device, class_names):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = outputs.max(1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Create confusion matrix
    cm = confusion_matrix(all_labels, all_preds)

    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.show()

    # Print classification report
    report = classification_report(all_labels, all_preds, target_names=class_names)
    print("Classification Report:")
    print(report)

    return cm, report

# Main function to run training
def main():
    # Set device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Data transforms
    data_transforms = {
        'train': transforms.Compose([
            transforms.RandomResizedCrop(224),
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(15),
            transforms.ColorJitter(brightness=0.1, contrast=0.1),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
        'val': transforms.Compose([
            transforms.Resize(256),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
    }

    # Dataset paths - updated for your specific directory structure
    train_path = "/content/Tomato_5_class/Tomato_5_class/Training_set"
    test_path = "/content/Tomato_5_class/Tomato_5_class/Testing_set"

    # Create datasets
    datasets = {
        'train': TomatoDataset(train_path, data_transforms['train']),
        'val': TomatoDataset(test_path, data_transforms['val'])
    }

    # Get number of classes
    num_classes = len(datasets['train'].classes)
    print(f"Number of classes: {num_classes}")
    print(f"Classes: {datasets['train'].classes}")

    # Display class distribution
    class_counts = datasets['train'].get_class_counts()
    print("\nClass distribution in training set:")
    for cls, count in class_counts.items():
        print(f"Class {cls}: {count} images")

    # Create weighted sampler to handle class imbalance
    train_sampler = create_weighted_sampler(datasets['train'])

    # Create dataloaders with weighted sampler for training
    dataloaders = {
        'train': DataLoader(datasets['train'], batch_size=16, sampler=train_sampler, num_workers=4),
        'val': DataLoader(datasets['val'], batch_size=16, shuffle=False, num_workers=4)
    }

    # Create model
    model = TomatoViT(num_classes=num_classes)

    # Loss function options:
    # 1. Weighted Cross Entropy Loss
    class_counts = Counter(datasets['train'].labels)
    class_weights = torch.FloatTensor([1.0 / class_counts[i] for i in range(num_classes)])
    class_weights = class_weights / class_weights.sum() * num_classes  # Normalize
    class_weights = class_weights.to(device)
    weighted_criterion = nn.CrossEntropyLoss(weight=class_weights)

    # 2. Focal Loss (alternative)
    focal_criterion = FocalLoss(gamma=2.0)

    # Choose which loss to use
    criterion = focal_criterion  # You can switch between weighted_criterion and focal_criterion

    # Optimizer with learning rate scheduling
    optimizer = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=0.05)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)

    # Training
    history = train_model(model, dataloaders, criterion, optimizer, device, num_epochs=20)

    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train Accuracy')
    plt.plot(history['val_acc'], label='Val Accuracy')
    plt.title('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

    # Save model
    torch.save(model.state_dict(), 'tomato_vit_model.pth')
    print("Model saved successfully!")

    # Evaluate model with confusion matrix
    evaluate_model(model, dataloaders['val'], device, datasets['train'].classes)

if __name__ == "__main__":
    main()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from transformers import ViTFeatureExtractor, TFViTForImageClassification
import numpy as np

# Set mixed precision policy
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Dataset paths
train_path = "/content/Tomato_5_class/Tomato_5_class/Training_set"
test_path = "/content/Tomato_5_class/Tomato_5_class/Testing_set"

# Initialize ViT feature extractor
vit_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224-in21k')

# Custom preprocessing function with proper normalization and shape handling
def vit_preprocess(image):
    # Convert to NumPy array if it's not already
    if tf.is_tensor(image):
        image = image.numpy()

    # Process with ViT feature extractor
    processed = vit_extractor(images=image, return_tensors='tf')['pixel_values']
    # Convert from shape [1, 3, 224, 224] to [224, 224, 3]
    processed = tf.transpose(tf.squeeze(processed, axis=0), [1, 2, 0])
    return processed

# EfficientNet preprocessing function
def efficientnet_preprocess(image):
    # EfficientNet expects values in [-1, 1]
    return tf.keras.applications.efficientnet.preprocess_input(image)

# Data augmentation
train_datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator()

# Data generators
train_gen = train_datagen.flow_from_directory(
    train_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=True
)

test_gen = test_datagen.flow_from_directory(
    test_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=False
)

# Hybrid Model Architecture
class ViTLayer(layers.Layer):
    def __init__(self):
        super().__init__()
        self.vit = TFViTForImageClassification.from_pretrained('google/vit-base-patch16-224-in21k').vit

    def call(self, inputs):
        # Using tf.transpose instead of iterating through batch elements
        # Convert from [batch, height, width, channels] to [batch, channels, height, width]
        x = tf.transpose(inputs, [0, 3, 1, 2])

        # The ViT expects pixel values normalized according to the model's requirements
        # We'll rely on the model to handle normalization internally
        return self.vit(x).last_hidden_state[:, 0]  # CLS token

def create_hybrid_model(num_classes=5):
    # Input layer
    inputs = layers.Input(shape=(224, 224, 3))

    # CNN Branch
    effnet_input = tf.keras.applications.efficientnet.preprocess_input(inputs)
    cnn = tf.keras.applications.EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_tensor=effnet_input
    )
    # Freeze the EfficientNet layers
    for layer in cnn.layers:
        layer.trainable = False

    cnn_features = layers.GlobalAveragePooling2D()(cnn.output)
    cnn_features = layers.Dense(256, activation='swish')(cnn_features)

    # ViT Branch - directly load the model
    vit_model = TFViTForImageClassification.from_pretrained(
        'google/vit-base-patch16-224-in21k',
        output_hidden_states=True
    )
    # Freeze ViT layers
    for layer in vit_model.layers:
        layer.trainable = False

    # We need to transpose input for ViT
    vit_input = layers.Lambda(lambda x: tf.transpose(x, [0, 3, 1, 2]))(inputs)
    vit_output = vit_model(vit_input, training=False).hidden_states[-1][:, 0]
    vit_features = layers.Dense(256, activation='swish')(vit_output)

    # Feature Fusion
    combined = layers.Concatenate()([cnn_features, vit_features])
    combined = layers.BatchNormalization()(combined)
    combined = layers.Dropout(0.5)(combined)

    # Classifier
    outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(combined)

    return models.Model(inputs=inputs, outputs=outputs)

# Focal Loss
class FocalLoss(tf.keras.losses.Loss):
    def __init__(self, alpha=0.25, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def call(self, y_true, y_pred):
        ce = tf.keras.losses.sparse_categorical_crossentropy(y_true, y_pred, from_logits=False)
        pt = tf.exp(-ce)
        return tf.reduce_mean(self.alpha * (1-pt)**self.gamma * ce)

# Learning Rate Schedule
def lr_schedule(epoch):
    initial_lr = 3e-4
    if epoch > 20:
        return initial_lr * 0.1
    elif epoch > 10:
        return initial_lr * 0.5
    return initial_lr

# Model Compilation
model = create_hybrid_model()
model.compile(
    optimizer=optimizers.AdamW(learning_rate=3e-4),
    loss=FocalLoss(),
    metrics=['accuracy']
)

# Class weights
class_counts = np.bincount(train_gen.classes)
total = len(train_gen.classes)
class_weights = {i: total/(len(class_counts)*count) for i, count in enumerate(class_counts)}

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3),
    ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy'),
    tf.keras.callbacks.LearningRateScheduler(lr_schedule)
]

# Alternative approach if you still face issues:
# Create a data pipeline using tf.data instead of ImageDataGenerator
def create_dataset(directory, batch_size=32, is_training=True):
    def preprocess(image, label):
        # Resize image
        image = tf.image.resize(image, [224, 224])
        if is_training:
            # Apply augmentation
            image = tf.image.random_flip_left_right(image)
            image = tf.image.random_brightness(image, 0.2)
            # Add more augmentations as needed
        return image, label

    # Create dataset from directory
    dataset = tf.keras.preprocessing.image_dataset_from_directory(
        directory,
        batch_size=batch_size,
        image_size=(224, 224),
        shuffle=is_training
    )

    # Apply preprocessing
    dataset = dataset.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    return dataset

# Uncomment these lines to use tf.data instead of ImageDataGenerator
# train_data = create_dataset(train_path, batch_size=32, is_training=True)
# test_data = create_dataset(test_path, batch_size=32, is_training=False)

# Training
try:
    # Try with ImageDataGenerator
    history = model.fit(
        train_gen,
        validation_data=test_gen,
        epochs=30,
        class_weight=class_weights,
        callbacks=callbacks
    )
except Exception as e:
    print(f"Error with ImageDataGenerator: {e}")
    print("Falling back to tf.data approach...")

    # Uncomment to use tf.data approach if ImageDataGenerator fails
    # train_data = create_dataset(train_path, batch_size=32, is_training=True)
    # test_data = create_dataset(test_path, batch_size=32, is_training=False)
    # history = model.fit(
    #     train_data,
    #     validation_data=test_data,
    #     epochs=30,
    #     callbacks=callbacks
    # )

# Evaluation
try:
    model.load_weights('best_model.h5')
    test_loss, test_acc = model.evaluate(test_gen)
    print(f"\nTest Accuracy: {test_acc:.4f}")
    print(f"Test Loss: {test_loss:.4f}")
except Exception as e:
    print(f"Error during evaluation: {e}")
```

```python
# ✅ STEP 1: Mount Google Drive & Extract Dataset
from google.colab import drive
import zipfile
import os

drive.mount('/content/drive')

zip_path = "/content/drive/MyDrive/Tomato Maturity Detection Dataset.zip"
extract_path = "/content/tomato_maturity"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# ✅ STEP 2: Import Libraries
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.optim import AdamW
from transformers import ViTForImageClassification
from tqdm import tqdm
import numpy as np
import random

# ✅ STEP 3: Set Seed and Device
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ✅ STEP 4: Data Transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5]*3, std=[0.5]*3)
])

# ✅ STEP 5: Load Dataset and Split
data_dir = os.path.join(extract_path, "Tomato Maturity Detection Dataset", "Augment Dataset")
full_dataset = ImageFolder(root=data_dir, transform=transform)

train_size = int(0.7 * len(full_dataset))
val_size = int(0.15 * len(full_dataset))
test_size = len(full_dataset) - train_size - val_size

train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)

# ✅ STEP 6: Define ViT Model for 2 Classes
model = ViTForImageClassification.from_pretrained(
    "google/vit-base-patch16-224-in21k",
    num_labels=2  # Binary: MATURE, IMMATURE
)
model.to(device)

# ✅ STEP 7: Training Config
optimizer = AdamW(model.parameters(), lr=2e-5)
criterion = nn.CrossEntropyLoss()

best_val_acc = 0.0
save_path = "best_vit_tomato_model.pth"

# ✅ STEP 8: Training and Evaluation Functions
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, correct, total = 0, 0, 0

    for imgs, labels in tqdm(loader, desc="Training"):
        imgs, labels = imgs.to(device), labels.to(device)
        outputs = model(imgs).logits
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / len(loader), correct / total

def evaluate(model, loader, criterion):
    model.eval()
    total_loss, correct, total = 0, 0, 0

    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Evaluating"):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs).logits
            loss = criterion(outputs, labels)

            total_loss += loss.item()
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    return total_loss / len(loader), correct / total

# ✅ STEP 9: Training Loop
EPOCHS = 5
for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion)
    print(f"Train Loss: {train_loss:.4f} | Train Accuracy: {train_acc:.4f}")

    val_loss, val_acc = evaluate(model, val_loader, criterion)
    print(f"Val Loss: {val_loss:.4f} | Val Accuracy: {val_acc:.4f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), save_path)
        print(f"✅ Best model saved with val acc: {val_acc:.4f}")

# ✅ STEP 10: Final Test
model.load_state_dict(torch.load(save_path))
model.eval()

test_loss, test_acc = evaluate(model, test_loader, criterion)
print(f"\n📊 Final Test Accuracy: {test_acc:.4f} | Loss: {test_loss:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import numpy as np

# Set mixed precision policy
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Dataset paths
train_path = "/content/Tomato_5_class/Tomato_5_class/Training_set"
test_path = "/content/Tomato_5_class/Tomato_5_class/Testing_set"

# Data augmentation
train_datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest',
    preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
)

test_datagen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
)

# Data generators
train_gen = train_datagen.flow_from_directory(
    train_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=True
)

test_gen = test_datagen.flow_from_directory(
    test_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='sparse',
    shuffle=False
)

# Create a simplified hybrid model
def create_enhanced_model(num_classes=5):
    # Input layer
    inputs = layers.Input(shape=(224, 224, 3))

    # CNN Branch 1: EfficientNetB0
    effnet = tf.keras.applications.EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_tensor=inputs
    )

    # Freeze early layers
    for layer in effnet.layers[:-20]:  # Keep some later layers trainable
        layer.trainable = False

    effnet_features = layers.GlobalAveragePooling2D()(effnet.output)

    # CNN Branch 2: MobileNetV3 (as a alternative to ViT)
    mobilenet = tf.keras.applications.MobileNetV3Large(
        include_top=False,
        weights='imagenet',
        input_tensor=inputs
    )

    # Freeze early layers
    for layer in mobilenet.layers[:-20]:
        layer.trainable = False

    mobilenet_features = layers.GlobalAveragePooling2D()(mobilenet.output)

    # Feature fusion
    effnet_features = layers.Dense(256, activation='swish')(effnet_features)
    mobilenet_features = layers.Dense(256, activation='swish')(mobilenet_features)

    combined = layers.Concatenate()([effnet_features, mobilenet_features])
    combined = layers.BatchNormalization()(combined)
    combined = layers.Dropout(0.5)(combined)

    # Add attention mechanism (self-attention)
    attention = layers.Dense(512, activation='tanh')(combined)
    attention = layers.Dense(1, activation='sigmoid')(attention)
    combined = layers.Multiply()([combined, attention])

    # Classifier with deeper layers
    x = layers.Dense(512, activation='swish')(combined)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(256, activation='swish')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    # Final classifier
    outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(x)

    return models.Model(inputs=inputs, outputs=outputs)

# Focal Loss
class FocalLoss(tf.keras.losses.Loss):
    def __init__(self, alpha=0.25, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def call(self, y_true, y_pred):
        ce = tf.keras.losses.sparse_categorical_crossentropy(y_true, y_pred, from_logits=False)
        pt = tf.exp(-ce)
        return tf.reduce_mean(self.alpha * (1-pt)**self.gamma * ce)

# Learning Rate Schedule
def lr_schedule(epoch):
    initial_lr = 3e-4
    if epoch > 20:
        return initial_lr * 0.1
    elif epoch > 10:
        return initial_lr * 0.5
    return initial_lr

# Model Compilation
model = create_enhanced_model()
model.compile(
    optimizer=optimizers.AdamW(learning_rate=3e-4, weight_decay=1e-5),
    loss=FocalLoss(),
    metrics=['accuracy']
)

# Class weights
class_counts = np.bincount(train_gen.classes)
total = len(train_gen.classes)
class_weights = {i: total/(len(class_counts)*count) for i, count in enumerate(class_counts)}

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3),
    ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy'),
    tf.keras.callbacks.LearningRateScheduler(lr_schedule)
]

# Training
history = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=30,
    class_weight=class_weights,
    callbacks=callbacks
)

# Evaluation
model.load_weights('best_model.h5')
test_loss, test_acc = model.evaluate(test_gen)
print(f"\nTest Accuracy: {test_acc:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
import os
import torch
from torchvision import datasets
from torch.utils.data import DataLoader, random_split
from transformers import ViTImageProcessor
from PIL import Image

# Constants
IMG_SIZE = 224
BATCH_SIZE = 32
SEED = 42
train_dir = '/content/Tomato_5_class/Tomato_5_class/Training_set'
test_dir = '/content/Tomato_5_class/Tomato_5_class/Testing_set'

# Load image processor (replaces ViTFeatureExtractor)
image_processor = ViTImageProcessor.from_pretrained("google/vit-base-patch16-224-in21k")

# Custom transform using image processor
def hf_transform(img):
    return image_processor(images=img, return_tensors="pt")["pixel_values"].squeeze(0)

# Custom dataset wrapper
class TomatoDataset(torch.utils.data.Dataset):
    def __init__(self, dataset):
        self.dataset = dataset

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        img = hf_transform(img)
        return img, label

# Load raw dataset from folders
full_dataset = datasets.ImageFolder(root=train_dir)

# Split: 80% train, 20% validation
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_ds, val_ds = random_split(full_dataset, [train_size, val_size], generator=torch.Generator().manual_seed(SEED))

# Wrap in custom dataset with Hugging Face transform
train_dataset = TomatoDataset(train_ds)
val_dataset = TomatoDataset(val_ds)

# Load test set
test_raw_dataset = datasets.ImageFolder(root=test_dir)
test_dataset = TomatoDataset(test_raw_dataset)

# Dataloaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)
```
